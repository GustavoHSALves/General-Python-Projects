### Criando o banco de dados

```python
# Reemove o arquivo com o banco de dados SQLite (caso exista)
import os
os.remove("nerddosdados.db") if os.path.exists("nerddosdados.db") else None
```

```python
# Importando o módulo de acesso ao SQLite
#!pip install pysqlite
import sqlite3
```

```python
# Cria uma conexão com o banco de dados. 
# Se o banco de dados não existir, ele é criado neste momento.
conexao = sqlite3.connect('nerddosdados.db')
```

```python
# Verificando o tipo do objeto, podemos observar que é um objeto do SQLite Connection
type(conexao)
```

```python
# Criando um cursor 
# (Um cursor permite percorrer todos os registros em um conjunto de dados)
cursor = conexao.cursor()
```

```python
# Verificando o tipo do objeto, podemos observar que é um objeto do SQLite Cursor
type(cursor)
```

```python
# Cria uma instrução/comando sql
comando = 'CREATE TABLE VIDEOS (ID INTEGER PRIMARY KEY, TITULO VARCHAR(100), CATEGORIA VARCHAR(140))'
```

```python
type(comando)
```

```python
# Executando a instrução sql no cursor
cursor.execute(comando)
```

# Existem 2 formas de fazer o insert

```python
# Forma numero 1 - fazendo um insert por vez
cursor.execute("INSERT INTO VIDEOS VALUES(1, 'Banco de Dados SQLite', 'Banco de Dados')")
conexao.commit()
#cursor.close()
#conexao.close()
```

```python
# Forma numero 2 - fazendo um insert para inserir vários registros de uma vez
comando = 'INSERT INTO VIDEOS VALUES (?, ?, ?)'
```

```python
type(comando)
```

```python
# Dados
registros = [(2, 'Como fazer Normalização e Padronização de Dados', 'Machine Learning'),
             (3, 'Como prever vendas com Machine Learning', 'Machine Learning'),
             (4, 'Como aprender Power BI do Zero', 'Visualização de Dados')]
```

```python
type(registros)
```

```python
# Inserindo os registros
for registro in registros:    
    cursor.execute(comando, registro)
    
# Grava a transação
conexao.commit()
```

```python
# Criando outra sentença SQL para selecionar registros
comando = 'select * from videos'
```

```python
# Seleciona todos os registros e recupera os registros
cursor.execute(comando)
dados = cursor.fetchall()
```

```python
# Exibindo os registro do banco de dados
for linha in dados:
    print ('ID_VIDEO: %d, Título: %s, Categoria: %s \n' % linha)
```

```python
# Leitura de registros específicos com condições
def leitura_registros():
    cursor.execute("SELECT * FROM VIDEOS WHERE CATEGORIA = 'Machine Learning'")
    for linha in cursor.fetchall():
        print(linha) 
```

```python
leitura_registros()
```

```python
# Update - Atualizando registro
cursor.execute("UPDATE VIDEOS SET TITULO = 'Prevendo Vendas com Machine Learning' WHERE ID = 3")
conexao.commit()
```

```python
leitura_registros()
```

```python
# Delete - Excluindo registro
cursor.execute("DELETE FROM VIDEOS WHERE ID = 3")
conexao.commit()
```

```python
leitura_registros()
```
